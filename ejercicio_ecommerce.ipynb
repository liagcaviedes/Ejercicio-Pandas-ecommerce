¿Cuáles fueron los cinco *email hosts* más populares (e.g., gmail, yahoo,...)?

````python
df['Email'].apply(lambda x: x.split('@')[1]).head(5)
````

````python
# 1. Extraer el host del email (lo que está después de '@') 
df['email_host'] = df['Email'].str.split('@').str[1]

# 2. Contar las ocurrencias de cada host
email_host_counts = df['email_host'].value_counts()

# 3. Obtener los cinco hosts más comunes
top_5_hosts = email_host_counts.head(5)

print(top_5_hosts)
````
